## Data Analysis

**Which categories are the most profitable and best-selling?**

```python
df_category = pd.DataFrame(df.groupby(['Category'])[['Quantity', 'Sales', 'Profit']].sum().reset_index())
df_category
```

**Which sub-categories are the most profitable and best-selling?**

```python
df_subcat = pd.DataFrame(df.groupby(['Sub-Category'])[['Quantity', 'Sales', 'Profit']].sum())

# Most Quantity ofsales
df_subcat.sort_values('Quantity', ascending = False).reset_index()
```

```python
# Best Selling
df_subcat.sort_values('Sales', ascending = False).reset_index()
```

```python
# Most Profitable
df_subcat.sort_values('Profit', ascending = False).reset_index()
```

 **Which segment of customers generates the most profit?**

```python
df_segment = pd.DataFrame(df.groupby(['Segment'])[['Profit']].sum().reset_index())
df_segment
```

**Which cities are the top 10 sales volume greatest?**

```python
df_city = pd.DataFrame(df.groupby(['City'])[['Sales', 'Quantity']].sum().sort_values('Sales',ascending = False).reset_index())
df_city.head(10)
```

**Who are the top 20 profit-generating customers?**

```python
# Calculate the profit grouped by Customer Name
df_customer = pd.DataFrame(df.groupby('Customer Name')['Profit'].sum())

# Order top 10 customers who make the most profit
df_customer = pd.DataFrame(df_customer.sort_values('Profit', ascending = False).round(2).reset_index().head(20))
df_customer
```
